```python
from sklearn.model_selection import train_test_split
```

```python
y=data['quality_index']
x=data.drop("quality_index",axis=1)
x=x.drop("quality",axis=1)
x.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
model=RandomForestClassifier(n_estimators=200)
model.fit(X_train,y_train)
Predictions=model.predict(X_test)
print(classification_report(y_test, Predictions))
model.score(X_test, y_test)
```

```python
model2=DecisionTreeClassifier()
model2.fit(X_train,y_train)
Predictions=model2.predict(X_test)
print(classification_report(y_test, Predictions))
model2.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model3=KNeighborsClassifier(n_neighbors=1)
model3.fit(X_train,y_train)
Predictions=model3.predict(X_test)
print(classification_report(y_test, Predictions))
model3.score(X_test, y_test)
```

```python
from tensorflow import keras
import tensorflow as tf
from keras import backend as K
from keras.layers import Dense
model = tf.keras.Sequential()
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(11, activation='relu',input_dim=11))




model.add(tf.keras.layers.Dense(22, activation='relu'))



model.add(tf.keras.layers.Dense(22, activation='relu'))


# Add a softmax layer with 10 output units:

model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python

model.compile(optimizer=tf.train.AdamOptimizer(), 

              loss='binary_crossentropy',

              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=200)
```

```python
test_loss, test_acc = model.evaluate(X_train, y_train)
```
